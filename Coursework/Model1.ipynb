```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
df = pd.read_csv('../Data/dataset1_scaled.csv', index_col = 0)
df.head()
```

```python
y = pd.read_csv('../Data/dataset1.csv', index_col = 0)
y = y[y.columns[-1]]
y.head()
```

```python
# import keras stuff
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.utils import to_categorical
```

```python
X = df
y_binary = to_categorical(y)
X_train, X_test, y_train, y_test = train_test_split(X,y_binary, test_size = 0.8, random_state = 23)
```

```python
model = Sequential()
model.add(
    Dense(input_shape = (6,), units = 10, activation = 'relu')
) # add first dense layer
model.add(
    Dense(units = 10, activation = 'sigmoid')
)
model.add(
    Dense(units = 2, activation = 'sigmoid')
)

#model.add(
#    BatchNormalization()
#)
model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'sgd',
    metrics = ['accuracy']
)
```

```python
model.fit(X_train,y_train, epochs = 100, batch_size = 32)
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print(f'Test loss: {score[0]} / Test accuracy: {score[1]}')
```

Adding code to make this process smoother...

```python
import tensorflow as tf

tf.random.set_seed(1)

from tensorflow.keras import Input, Model
def uncompiled_model():
    inputs =  Input(shape = (6,), name = 'Data')
    x = Dense(16, activation = 'relu', name = 'dense_1')(inputs)
    x = BatchNormalization()(x)
    x = Dense(16, activation = 'relu', name = 'dense_2')(x)

    outputs = Dense(2, activation = 'softmax', name = 'TargetHit')(x)
    model = Model(inputs = inputs, outputs = outputs)
    return model
    
    
def compile_model():
    model = uncompiled_model()
    model.compile(
        optimizer="sgd",
        loss="categorical_crossentropy",
        metrics=["accuracy",
                 "binary_accuracy",
                 "binary_crossentropy",
                 "categorical_accuracy"
                ],
    )
    return model
```

```python


model = compile_model()
model.fit(X_train, y_train, epochs = 100, verbose = 0)
score = model.evaluate(X_test, y_test, verbose=1)
score
```
